```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import math


df1=pd.read_csv("ChampsElysées.csv",sep=";")
df2=pd.read_csv("Convention.csv",sep=";")
df3=pd.read_csv("St_peres.csv",sep=";")

df_meteo=pd.read_csv("donnees-synop-essentielles-omm.csv",sep=";",encoding='unicode_escape')

df_data=pd.concat([df1,df2,df3],ignore_index=True)
df_data.head()
```

```python
df_meteo.sort_values("Date")
```

```python
df_train_meteo=pd.DataFrame()

df_train_meteo["Date"]=df_meteo["Date"]
df_train_meteo["Temp"]=df_meteo["Temprature"]
df_train_meteo["Pluie 1h"]=df_meteo["Prcipitations dans la dernire heure"]
df_train_meteo["Pluie 3h"]=df_meteo["Prcipitations dans les 3 dernires heures"]
df_train_meteo["Neige"]=df_meteo["Hauteur totale de la couche de neige, glace, autre au sol"]
df_train_meteo["Visibilité Horiz"]=df_meteo["Visibilit horizontale"]
df_train_meteo=pd.concat([df_train_meteo,df_train_meteo,df_train_meteo],ignore_index=True)
ls_index=df_train_meteo.sort_values(by="Date").index

for index in range(0,len(ls_index)-1,3):
    datetime_object = datetime.strptime(df_train_meteo["Date"][ls_index[index]], '%Y-%m-%dT%H:%M:%S%z')
    formatted_datetime_1=(datetime_object + timedelta(hours=1)).strftime('%Y-%m-%dT%H:%M:%S%z')
    formatted_datetime_2=(datetime_object + timedelta(hours=2)).strftime('%Y-%m-%dT%H:%M:%S%z')

    df_train_meteo["Date"][ls_index[index+1]]=f"{formatted_datetime_1[:-2]}:{formatted_datetime_1[-2:]}"
    df_train_meteo["Date"][ls_index[index+2]]=f"{formatted_datetime_2[:-2]}:{formatted_datetime_2[-2:]}"



df_train_meteo
```

```python
df_train_meteo.sort_values("Date")
```

```python
df_tot=pd.merge(left=df_data,right=df_train_meteo, left_on="Date et heure de comptage", right_on="Date")
df_tot
```

```python
df_tot.info()
```

Taux d’occupation (en pourcentage de temps d'occupation de la station de mesure par des véhicules sur l’heure).
Selon la valeur de K, on peut qualifier l’état de trafic au point de mesure :
- Fuide : 1
- Pré-saturé : 2
- Saturé : 3
- Bloqué : 4
- Inconnu: 0

État ouvert ou non (barré, inconnu ou invalide) à la circulation de l’arc :
- Inconnu : 0
- Ouvert : 1
- Barré : 2
- Invalide : 3

```python
etat_trafic={"Fluide":1,"Pré-saturé":2,"Saturé":3,"Bloqué":4,"Inconnu":0}
etat_arc={"Ouvert":1,"Barré":2,"Invalide":3,"Inconnu":0}
```

```python
df_train=pd.DataFrame(data={"id arc":df_tot["Identifiant arc"], "id noeud amont": df_tot["Identifiant noeud amont"],"id noeud aval":df_tot["Identifiant noeud aval"],"débit horaire":df_tot["Débit horaire"],"taux occupation":df_tot["Taux d\'occupation"],"Temp":df_tot["Temp"],"Pluie 1h":df_tot["Pluie 1h"],"Pluie 3h":df_tot["Pluie 3h"],"Neige":df_tot["Neige"],"Visibilité Horiz":df_tot["Visibilité Horiz"]})
df_train["year"]=df_tot["Date et heure de comptage"].apply(lambda time: time[:4])
df_train["month"]=df_tot["Date et heure de comptage"].apply(lambda time: time[5:7])
df_train["day"]=df_tot["Date et heure de comptage"].apply(lambda time: time[8:10])
df_train["hour"]=df_tot["Date et heure de comptage"].apply(lambda time: time[11:13])
df_train["etat trafic"]=df_tot["Etat trafic"].apply(lambda etat:etat_trafic[etat])
df_train["etat arc"]=df_tot["Etat arc"].apply(lambda etat:etat_arc[etat])

df_train=df_train[["year","month","day","hour","id arc","id noeud amont","id noeud aval","etat trafic","etat arc","Temp","Pluie 1h","Pluie 3h","Neige","Visibilité Horiz","débit horaire","taux occupation"]]
df_train=df_train.dropna()

df_train
```

```python
df_train.sort_values("year")
```

```python
df_train['hour'] = df_train['hour'].astype(int)
df_train['month'] = df_train['month'].astype(int)
df_train['day'] = df_train['day'].astype(int)

df_train["hour_norm"] = 2 * math.pi * df_train["hour"] / df_train["hour"].max()

df_train["cos_hour"] = np.cos(df_train["hour_norm"])

df_train["sin_hour"] = np.sin(df_train["hour_norm"])

df_train["day_norm"] = 2 * math.pi * df_train["day"] / df_train["day"].max()

df_train["cos_day"] = np.cos(df_train["day_norm"])

df_train["sin_day"] = np.sin(df_train["day_norm"])

df_train["month_norm"] = 2 * math.pi * df_train["month"] / df_train["month"].max()

df_train["cos_month"] = np.cos(df_train["month_norm"])

df_train["sin_month"] = np.sin(df_train["month_norm"])

```

```python
df_train.to_csv('Train.csv')
df_train[df_train["id arc"]==191].sort_values(by=["year","month","day","hour"])
```
